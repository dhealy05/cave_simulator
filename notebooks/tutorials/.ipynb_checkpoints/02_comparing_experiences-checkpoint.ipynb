# Tutorial 2: Comparing Experiences

This notebook is an API-first walkthrough of Cave comparison workflows. Tutorial 1 built one subjective trajectory. Here we compare many trajectories.

The central rule is:

```text
comparison is over trajectories and readouts, not over screenshots
```

We will compare:

- the same external sequence across different subjects;
- the same subject across different experience sequences;
- raw memory versus a matched baseline effect;
- embedding components used by distance matrices;
- topology density surfaces and population trajectory geometry;
- report-like comparison bundles generated from package APIs.

## 0. Setup

Generated artifacts go under `out/tutorials/02_comparing_experiences/`.

````python
from __future__ import annotations

from dataclasses import replace
from pathlib import Path
import json
import math
import sys

import numpy as np
import matplotlib.pyplot as plt

try:
    from IPython.display import Image, Markdown, display
except Exception:
    def display(value):
        print(value)

    class Markdown(str):
        pass

    class Image:
        def __init__(self, filename: str, width: int | None = None):
            self.filename = filename
            self.width = width


def find_repo_root(start: Path) -> Path:
    for candidate in (start, *start.parents):
        if (candidate / "cave").exists() and (candidate / "notebooks").exists():
            return candidate
    return start

ROOT = find_repo_root(Path.cwd())
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

OUT = ROOT / "out" / "tutorials" / "02_comparing_experiences"
OUT.mkdir(parents=True, exist_ok=True)


def show_artifact(path: Path, width: int = 780) -> None:
    path = Path(path)
    if path.exists():
        display(Image(filename=str(path), width=width))
    else:
        display(Markdown(f"**Missing artifact:** `{path}`"))


def markdown_table(headers: list[str], rows: list[list[object]]) -> None:
    lines = ["| " + " | ".join(headers) + " |", "| " + " | ".join(["---"] * len(headers)) + " |"]
    for row in rows:
        lines.append("| " + " | ".join(str(item) for item in row) + " |")
    display(Markdown("\n".join(lines)))
````

## 1. Shared Building Blocks

The comparison APIs operate on `Episode`s. We will build a few sequences and subject configurations, then package them as `EpisodeSet` and `PopulationRunRecord` objects.

````python
from cave.commitments.attention import AttentionProfile, INTERNAL_EXPECTATION_CHANNEL
from cave.commitments.memory import MemoryParams, MemoryTrace
from cave.commitments.topology import SubjectiveTopologyParams, SubjectiveTopologyPrior
from cave.demonstrations.examples import DEFAULT_VOCABULARY, default_model_params, model_for_sequence
from cave.demonstrations.simulation import ExperienceModel
from cave.demonstrations.state import SubjectState
from cave.observation.episodes import CaveProducer
from cave.observation.episode_runs import episode_set, labeled_episode
from cave.observation.experience import ExperienceObject, FeatureProjection, FeatureVector, InputSequence, TemporalExtent
from cave.observation.experience.presentations import AudioPresentation
from cave.observation.population import PopulationRunRecord, factor_level
from cave.observation.sensing import Sensorium
from cave.observation.sensing.sensor import FeatureSensor, visual_feature_sensor

vocabulary = list(DEFAULT_VOCABULARY)

sensorium = Sensorium(
    sensors=(
        visual_feature_sensor(),
        FeatureSensor(modality="audio", channel="audio", gain=1.0),
    )
)

topology_params = SubjectiveTopologyParams(
    feature_x=FeatureProjection(
        name="form",
        weights={"angularity": 0.5, "symmetry": 0.25, "sides": 0.25},
    ),
    feature_y=FeatureProjection(
        name="sensory tone",
        weights={"roundness": 0.4, "hue": 0.2, "saturation": 0.2, "novelty": 0.2},
    ),
    prior=SubjectiveTopologyPrior(mode="random_wells", strength=0.12, width=0.28, seed=23, well_count=4),
    deposit_strength=0.22,
    expectation_deposit_strength=0.15,
)

base_params = replace(
    default_model_params(),
    memory=MemoryParams(retention=0.78, decay_tau=1.8, max_age=4.0),
    attention=AttentionProfile(
        mode="constant",
        level=0.78,
        channel_weights={"visual": 0.50, "audio": 0.12, INTERNAL_EXPECTATION_CHANNEL: 0.38},
    ),
    topology=topology_params,
)


def features(**kwargs: float) -> FeatureVector:
    values = {
        "sides": 0.0,
        "size": 0.5,
        "hue": 0.0,
        "saturation": 0.55,
        "lightness": 0.72,
        "angularity": 0.5,
        "roundness": 0.5,
        "symmetry": 0.5,
        "novelty": 0.3,
    }
    values.update(kwargs)
    return FeatureVector(values)


def obj(id: str, start: float, end: float, order: int, fv: FeatureVector, *, salience: float = 0.8, modality: str = "visual") -> ExperienceObject:
    presentation = AudioPresentation(label=id) if modality == "audio" else None
    return ExperienceObject(
        id=id,
        temporal_extent=TemporalExtent(start, end, order),
        features=fv,
        salience=salience,
        modality=modality,
        presentation=presentation,
    )


def make_sequence(treatment: str) -> InputSequence:
    if treatment == "sharp_then_round":
        return InputSequence(objects=[
            obj("sharp", 0.0, 1.0, 0, features(sides=0.0, angularity=1.0, roundness=0.0, symmetry=0.86, hue=0.12, novelty=0.6), salience=0.9),
            obj("tone", 0.65, 1.65, 1, features(angularity=0.2, roundness=0.7, hue=0.08, saturation=0.9, novelty=0.7), salience=0.7, modality="audio"),
            obj("round", 1.85, 2.85, 2, features(sides=1.0, angularity=0.0, roundness=1.0, symmetry=1.0, hue=0.58, novelty=0.35), salience=0.85),
            obj("rest", 3.15, 3.85, 3, features(size=0.2, angularity=0.05, roundness=0.08, novelty=0.1), salience=0.35),
        ])
    if treatment == "repeat_then_violate":
        return InputSequence(objects=[
            obj("repeat_a", 0.0, 0.8, 0, features(sides=0.0, angularity=0.95, roundness=0.05, symmetry=0.82, novelty=0.45), salience=0.82),
            obj("repeat_b", 1.0, 1.8, 1, features(sides=0.0, angularity=0.95, roundness=0.05, symmetry=0.82, novelty=0.25), salience=0.82),
            obj("repeat_c", 2.0, 2.8, 2, features(sides=0.0, angularity=0.95, roundness=0.05, symmetry=0.82, novelty=0.2), salience=0.82),
            obj("violation", 3.0, 3.8, 3, features(sides=1.0, angularity=0.0, roundness=1.0, symmetry=0.95, hue=0.62, novelty=0.9), salience=0.86),
        ])
    if treatment == "audio_overlap":
        return InputSequence(objects=[
            obj("visual_marker", 0.0, 1.1, 0, features(sides=0.05, angularity=0.85, roundness=0.15, symmetry=0.78, novelty=0.48), salience=0.8),
            obj("audio_marker", 0.0, 1.1, 1, features(angularity=0.15, roundness=0.75, hue=0.1, saturation=0.95, novelty=0.75), salience=0.82, modality="audio"),
            obj("quiet_circle", 1.6, 2.7, 2, features(sides=1.0, angularity=0.05, roundness=0.95, symmetry=1.0, hue=0.54, novelty=0.25), salience=0.62),
            obj("sharp_end", 3.0, 3.8, 3, features(sides=0.0, angularity=1.0, roundness=0.02, symmetry=0.8, novelty=0.55), salience=0.78),
        ])
    raise ValueError(treatment)


def run_episode(sequence: InputSequence, params=base_params, *, initial_memory: np.ndarray | None = None, dt: float = 0.2):
    if initial_memory is None:
        model = model_for_sequence(sequence, params=params, vocabulary=vocabulary)
        model.sensorium = sensorium
    else:
        trace = MemoryTrace(
            vector=np.asarray(initial_memory, dtype=float),
            retention=params.memory.retention,
            decay_tau=params.memory.decay_tau,
            max_age=params.memory.max_age,
        )
        model = ExperienceModel(
            sequence=sequence,
            subject_state=SubjectState.initial(trace, params.topology),
            params=params,
            vocabulary=list(vocabulary),
            sensorium=sensorium,
        )
    return CaveProducer(model).run(dt=dt)

reference_sequence = make_sequence("sharp_then_round")
reference_episode = run_episode(reference_sequence)

markdown_table(
    ["treatment", "duration", "objects"],
    [
        [name, f"{make_sequence(name).duration:.1f}", len(make_sequence(name).objects)]
        for name in ["sharp_then_round", "repeat_then_violate", "audio_overlap"]
    ],
)
````

## 2. Same World, Different Subjects

Here the external sequence is fixed. Only subject configuration changes. This is the simplest Cave comparison: same wall, different access and internal history.

````python
from cave.presentation.renderers.matplotlib_renderer import LayoutSpec, MatplotlibRenderer
from cave.observation.views import default_views
from cave.presentation.renderers.episode_set_dashboard import save_episode_set_dashboard, save_episode_set_distances_json
from cave.presentation.renderers.topology_surface_renderer import save_topology_state_surface

subject_variants = {
    "baseline": base_params,
    "low_capacity": replace(
        base_params,
        attention=AttentionProfile(
            mode="constant",
            level=0.24,
            channel_weights={"visual": 0.50, "audio": 0.12, INTERNAL_EXPECTATION_CHANNEL: 0.38},
        ),
    ),
    "external_heavy": replace(
        base_params,
        attention=AttentionProfile(
            mode="constant",
            level=0.82,
            channel_weights={"visual": 0.72, "audio": 0.18, INTERNAL_EXPECTATION_CHANNEL: 0.10},
        ),
    ),
    "internal_heavy": replace(
        base_params,
        attention=AttentionProfile(
            mode="constant",
            level=0.82,
            channel_weights={"visual": 0.20, "audio": 0.08, INTERNAL_EXPECTATION_CHANNEL: 0.72},
        ),
    ),
}

same_world_items = []
subject_surface_paths = []
for name, params in subject_variants.items():
    episode = run_episode(reference_sequence, params=params)
    same_world_items.append(labeled_episode(episode, id=name, label=name.replace("_", " "), group="subject"))
    surface = OUT / f"same_world_{name}_surface.png"
    save_topology_state_surface(episode, surface, resolution=20)
    subject_surface_paths.append((name, surface))

same_world_set = episode_set(
    same_world_items,
    id="same_world_different_subjects",
    title="Same World, Different Subjects",
    comparison_axis="subject configuration",
)

same_world_dashboard = OUT / "01_same_world_subject_dashboard.png"
same_world_distances = OUT / "01_same_world_distances.json"
save_episode_set_dashboard(same_world_set, same_world_dashboard, samples=32)
save_episode_set_distances_json(same_world_set, same_world_distances, samples=32)

renderer = MatplotlibRenderer(layout=LayoutSpec(columns=2, dpi=110))
baseline_gif = OUT / "02_same_world_baseline_multiview.gif"
low_capacity_gif = OUT / "03_same_world_low_capacity_multiview.gif"
renderer.save_animation(same_world_items[0].episode, default_views(), baseline_gif, dt=0.2, fps=6)
renderer.save_animation(same_world_items[1].episode, default_views(), low_capacity_gif, dt=0.2, fps=6)

markdown_table(
    ["artifact", "path"],
    [
        ["dashboard", same_world_dashboard.relative_to(ROOT)],
        ["distances", same_world_distances.relative_to(ROOT)],
        ["baseline multiview", baseline_gif.relative_to(ROOT)],
        ["low-capacity multiview", low_capacity_gif.relative_to(ROOT)],
    ],
)
````

## 3. Same Subject, Different Experiences

Now the subject is fixed and the treatment sequence changes. This asks how different worlds deform the same subject-side machinery.

````python
treatment_items = []
treatment_surface_paths = []
for treatment in ["sharp_then_round", "repeat_then_violate", "audio_overlap"]:
    episode = run_episode(make_sequence(treatment), params=base_params)
    treatment_items.append(labeled_episode(episode, id=treatment, label=treatment.replace("_", " "), group="treatment"))
    surface = OUT / f"experience_{treatment}_surface.png"
    save_topology_state_surface(episode, surface, resolution=20)
    treatment_surface_paths.append((treatment, surface))

experience_set = episode_set(
    treatment_items,
    id="same_subject_different_experiences",
    title="Same Subject, Different Experiences",
    comparison_axis="experience sequence",
)

experience_dashboard = OUT / "04_experience_dashboard.png"
experience_distances = OUT / "04_experience_distances.json"
save_episode_set_dashboard(experience_set, experience_dashboard, samples=32)
save_episode_set_distances_json(experience_set, experience_distances, samples=32)

markdown_table(
    ["treatment", "final memory norm", "mean surprise", "surface"],
    [
        [
            item.id,
            f"{np.linalg.norm(item.episode.observations[-1].memory_state):.3f}",
            f"{np.mean([obs.surprise for obs in item.episode.observations]):.3f}",
            path.relative_to(ROOT),
        ]
        for item, (_, path) in zip(treatment_items, treatment_surface_paths)
    ],
)
````

````python
def plot_surprise_by_episode(items, output: Path) -> Path:
    fig, ax = plt.subplots(figsize=(8.4, 3.8))
    for item in items:
        times = [obs.t for obs in item.episode.observations]
        surprise = [obs.surprise for obs in item.episode.observations]
        ax.plot(times, surprise, linewidth=2.0, label=item.display_label)
    ax.set_title("Surprise by experience sequence")
    ax.set_xlabel("time")
    ax.set_ylabel("surprise")
    ax.grid(True, alpha=0.25)
    ax.legend(frameon=False, fontsize=8)
    fig.tight_layout()
    fig.savefig(output, dpi=140)
    plt.close(fig)
    return output

surprise_png = plot_surprise_by_episode(treatment_items, OUT / "05_surprise_by_experience.png")
````

## 4. Observed Memory Versus State Effect

Observed memory is the raw memory trajectory. A state effect subtracts a matched baseline. In the default comparison helper the baseline is the no-input memory baseline, but the same idea applies to any matched control trajectory.

Here we build a subject with an initial prior memory, then compare the input run against a matched blank run.

````python
from cave.observation.structural import memory_trajectory, resample_trajectory

prior = np.full(len(vocabulary), 0.28, dtype=float)
prior[vocabulary.index("angularity")] = 0.85
prior[vocabulary.index("roundness")] = 0.12
prior[vocabulary.index("symmetry")] = 0.78

blank_sequence = InputSequence(objects=[
    obj("blank", 0.0, reference_sequence.duration, 0, features(size=0.0, saturation=0.0, lightness=0.0, novelty=0.0), salience=0.0),
])

prior_input_episode = run_episode(reference_sequence, params=base_params, initial_memory=prior)
prior_blank_episode = run_episode(blank_sequence, params=base_params, initial_memory=prior)

observed = resample_trajectory(memory_trajectory(prior_input_episode), samples=48)
baseline = resample_trajectory(memory_trajectory(prior_blank_episode), samples=48)
effect = observed - baseline


def plot_state_effect_components(output: Path) -> Path:
    axis_indices = [vocabulary.index(name) for name in ["angularity", "roundness", "novelty"]]
    labels = [vocabulary[index] for index in axis_indices]
    x = np.linspace(0.0, 1.0, observed.shape[0])
    fig, axes = plt.subplots(3, 1, figsize=(8.2, 7.0), sharex=True)
    for dim, label in zip(axis_indices, labels):
        axes[0].plot(x, observed[:, dim], label=label)
        axes[1].plot(x, baseline[:, dim], label=label)
        axes[2].plot(x, effect[:, dim], label=label)
    axes[0].set_title("Observed memory")
    axes[1].set_title("Matched blank baseline")
    axes[2].set_title("State effect: observed - baseline")
    axes[2].set_xlabel("normalized time")
    for ax in axes:
        ax.set_ylabel("value")
        ax.grid(True, alpha=0.25)
        ax.legend(frameon=False, fontsize=8, loc="upper right")
    fig.tight_layout()
    fig.savefig(output, dpi=140)
    plt.close(fig)
    return output

state_effect_png = plot_state_effect_components(OUT / "06_state_effect_subtraction.png")
show_artifact(state_effect_png, width=820)
````

## 5. Subjective Trajectory Embedding

The current `subjective_trajectory_embedding` is a concatenation of resampled expected input, prediction error, surprise, and learning rate. This section makes that construction visible.

````python
from cave.observation.structural import (
    expected_trajectory_embedding,
    prediction_error_trajectory_embedding,
    surprise_learning_trajectory_embedding,
    subjective_trajectory_embedding,
)

embedding_episode = treatment_items[1].episode  # repeat_then_violate
expected = expected_trajectory_embedding(embedding_episode, samples=36)
error = prediction_error_trajectory_embedding(embedding_episode, samples=36)
surprise_learning = surprise_learning_trajectory_embedding(embedding_episode, samples=36)
subjective = subjective_trajectory_embedding(embedding_episode, samples=36)


def plot_embedding_construction(output: Path) -> Path:
    fig = plt.figure(figsize=(11.2, 7.4))
    grid = fig.add_gridspec(4, 1, height_ratios=(1.0, 1.0, 0.6, 0.45), hspace=0.48)
    axes = [fig.add_subplot(grid[index, 0]) for index in range(4)]
    im0 = axes[0].imshow(expected.T, aspect="auto", cmap="Blues", interpolation="nearest")
    axes[0].set_title("Resampled expected input")
    axes[0].set_yticks(range(len(vocabulary)))
    axes[0].set_yticklabels(vocabulary, fontsize=7)
    fig.colorbar(im0, ax=axes[0], fraction=0.02, pad=0.01)

    im1 = axes[1].imshow(error.T, aspect="auto", cmap="coolwarm", interpolation="nearest")
    axes[1].set_title("Resampled prediction error")
    axes[1].set_yticks(range(len(vocabulary)))
    axes[1].set_yticklabels(vocabulary, fontsize=7)
    fig.colorbar(im1, ax=axes[1], fraction=0.02, pad=0.01)

    axes[2].plot(surprise_learning[:, 0], label="surprise", color="#D55E00", linewidth=2.0)
    axes[2].plot(surprise_learning[:, 1], label="learning rate", color="#009E73", linewidth=2.0)
    axes[2].set_title("Surprise and learning-rate series")
    axes[2].grid(True, alpha=0.25)
    axes[2].legend(frameon=False, fontsize=8)

    axes[3].imshow(subjective.T, aspect="auto", cmap="viridis", interpolation="nearest")
    axes[3].set_title(f"Concatenated subjective trajectory embedding: {subjective.shape[0]} samples x {subjective.shape[1]} features")
    axes[3].set_yticks([])
    axes[3].set_xlabel("resampled time")
    fig.subplots_adjust(left=0.16, right=0.92, top=0.94, bottom=0.08, hspace=0.58)
    fig.savefig(output, dpi=140)
    plt.close(fig)
    return output

embedding_png = plot_embedding_construction(OUT / "07_embedding_construction.png")
show_artifact(embedding_png, width=980)
````

## 6. Population Records And Trajectory Geometry

`PopulationRunRecord` adds factor labels to episodes. Population renderers can then color and group runs by treatment, subject condition, start condition, or comparison role.

This compact population crosses three treatments with three subject conditions.

````python
from cave.presentation.renderers.topology_atlas_renderer import (
    save_topology_atlas,
    save_topology_atlas_metrics,
)
from cave.presentation.renderers.topology_population_renderer import (
    save_topology_population_dashboard,
    save_topology_scatter_migration,
)

population_records = []
for treatment in ["sharp_then_round", "repeat_then_violate", "audio_overlap"]:
    for condition, params in {
        "baseline": subject_variants["baseline"],
        "external_heavy": subject_variants["external_heavy"],
        "low_capacity": subject_variants["low_capacity"],
    }.items():
        episode = run_episode(make_sequence(treatment), params=params)
        record_id = f"{treatment}:{condition}"
        population_records.append(
            PopulationRunRecord(
                id=record_id,
                label=f"{treatment.replace('_', ' ')} / {condition.replace('_', ' ')}",
                episode=episode,
                factors={
                    "treatment": factor_level("treatment", treatment, label=treatment.replace("_", " ")),
                    "condition": factor_level("condition", condition, label=condition.replace("_", " ")),
                    "start_condition": factor_level("start_condition", "zero", label="zero initial state"),
                },
                comparison_role="tutorial_run",
                matched_set_id=treatment,
                group_id=condition,
            )
        )

population_set = episode_set(
    [record.to_labeled_episode(group_factor="condition", series_factor="treatment") for record in population_records],
    id="tutorial_population",
    title="Tutorial Population: Treatments x Subject Conditions",
    comparison_axis="treatment and condition",
)

population_dashboard = OUT / "08_population_episode_dashboard.png"
population_distances = OUT / "08_population_distances.json"
save_episode_set_dashboard(population_set, population_dashboard, samples=28)
save_episode_set_distances_json(population_set, population_distances, samples=28)

topology_population_png = OUT / "09_population_topology_dashboard.png"
save_topology_population_dashboard(
    population_records,
    topology_population_png,
    topology_params,
    title="Topology Population: Treatments x Subject Conditions",
    selected_time=2.4,
)

topology_atlas_png = OUT / "10_population_topology_atlas.png"
topology_atlas_metrics = OUT / "10_population_topology_atlas_metrics.json"
save_topology_atlas(population_set, topology_atlas_png, topology_params)
save_topology_atlas_metrics(population_set, topology_atlas_metrics, topology_params)

topology_migration_gif = OUT / "11_population_trajectory_geometry.gif"
save_topology_scatter_migration(
    population_records,
    topology_migration_gif,
    topology_params,
    title="Population Trajectory Geometry",
    color_factor="treatment",
    marker_factor="condition",
    point_kind="centroid",
    fps=5,
)

markdown_table(
    ["artifact", "path"],
    [
        ["episode dashboard", population_dashboard.relative_to(ROOT)],
        ["distance JSON", population_distances.relative_to(ROOT)],
        ["topology population dashboard", topology_population_png.relative_to(ROOT)],
        ["topology atlas", topology_atlas_png.relative_to(ROOT)],
        ["topology atlas metrics", topology_atlas_metrics.relative_to(ROOT)],
        ["trajectory geometry animation", topology_migration_gif.relative_to(ROOT)],
    ],
)
````

## 7. Reading Distance JSON

The dashboard image is convenient, but the numeric comparison payload is also written as JSON. It includes labels and distance matrices for each embedding.

````python
payload = json.loads(population_distances.read_text(encoding="utf-8"))
markdown_table(
    ["field", "value"],
    [
        ["episode set", payload["episode_set"]["id"]],
        ["samples", payload["samples"]],
        ["labels", len(payload["labels"])],
        ["distance matrices", ", ".join(payload["distances"].keys())],
    ],
)

for name, matrix in payload["distances"].items():
    arr = np.asarray(matrix, dtype=float)
    display(Markdown(f"**{name}** mean off-diagonal distance: `{arr[np.triu_indices_from(arr, k=1)].mean():.4f}`"))
````

## 8. What This Tutorial Establishes

The comparison stack has a few separable layers:

- `Episode`: one subject-side trajectory.
- `EpisodeSet`: labeled episodes for direct comparison.
- Embeddings: actual input, observed memory, state effect, subjective trajectory.
- Distance payloads and dashboards: numerical comparison readouts.
- `PopulationRunRecord`: factor-labeled episodes for treatment/condition/start designs.
- Topology population renderers: density, centroid paths, and trajectory migration in feature space.

The next tutorial can use the same comparison contract for CaveNet, pressure reports, and evolved recurrent subjects.
