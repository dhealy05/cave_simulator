# Tutorial 1: Intro To Cave And Subjective Trajectory

This notebook is an API-first walkthrough of the core Cave objects. It builds a small external sequence, configures a subject, runs the model, inspects the resulting `Episode`, and renders the same views used by the walkthrough report.

The central distinction is:

```text
external sequence -> subject transformation -> Episode -> view projections
```

The Plato connection is mediation: the subject does not get the world directly. It gets sensed signals through particular channels, attention limits, prior state, expectation, memory, value, and topology-like state. The rendered wall, timeline, memory, expectation bars, correction view, and topology surface are projections of that subject-side trajectory.

## 0. Setup

The notebook writes generated artifacts to `out/tutorials/01_intro_to_cave_subjective_trajectory/`. The helper functions below keep display code separate from model code.

````python
from __future__ import annotations

from dataclasses import replace
from pathlib import Path
import json
import math
import sys

import numpy as np
import matplotlib.pyplot as plt

try:
    from IPython.display import Image, Markdown, display
except Exception:  # lets this file run as a plain Python script for smoke checks
    def display(value):
        print(value)

    class Markdown(str):
        pass

    class Image:
        def __init__(self, filename: str, width: int | None = None):
            self.filename = filename
            self.width = width


def find_repo_root(start: Path) -> Path:
    for candidate in (start, *start.parents):
        if (candidate / "cave").exists() and (candidate / "notebooks").exists():
            return candidate
    return start

ROOT = find_repo_root(Path.cwd())
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

OUT = ROOT / "out" / "tutorials" / "01_intro_to_cave_subjective_trajectory"
OUT.mkdir(parents=True, exist_ok=True)


def show_artifact(path: Path, width: int = 760) -> None:
    path = Path(path)
    if path.exists():
        display(Image(filename=str(path), width=width))
    else:
        display(Markdown(f"**Missing artifact:** `{path}`"))


def markdown_table(headers: list[str], rows: list[list[object]]) -> None:
    lines = ["| " + " | ".join(headers) + " |", "| " + " | ".join(["---"] * len(headers)) + " |"]
    for row in rows:
        lines.append("| " + " | ".join(str(item) for item in row) + " |")
    display(Markdown("\n".join(lines)))
````

## 1. The External Sequence

An external Cave world is an `InputSequence` of temporally extended `ExperienceObject`s. Each object has features, a temporal extent, salience, a modality, and optional presentation metadata.

This first sequence includes visual objects and one audio object. Whether the audio object enters the subject depends on the subject's sensorium and attention channel weights.

````python
from cave.observation.experience import (
    ExperienceObject,
    FeatureVector,
    InputSequence,
    TemporalExtent,
)
from cave.observation.experience.presentations import AudioPresentation
from cave.demonstrations.examples import DEFAULT_VOCABULARY

vocabulary = list(DEFAULT_VOCABULARY)

sequence = InputSequence(
    objects=[
        ExperienceObject(
            id="sharp_triangle",
            temporal_extent=TemporalExtent(0.0, 1.1, 0),
            features=FeatureVector({
                "sides": 0.0,
                "size": 0.64,
                "hue": 0.12,
                "saturation": 0.72,
                "lightness": 0.72,
                "angularity": 1.0,
                "roundness": 0.0,
                "symmetry": 0.82,
                "novelty": 0.62,
            }),
            salience=0.92,
            modality="visual",
        ),
        ExperienceObject(
            id="warm_tone",
            temporal_extent=TemporalExtent(0.7, 1.8, 1),
            features=FeatureVector({
                "sides": 0.15,
                "size": 0.25,
                "hue": 0.08,
                "saturation": 0.9,
                "lightness": 0.55,
                "angularity": 0.2,
                "roundness": 0.65,
                "symmetry": 0.4,
                "novelty": 0.72,
            }),
            salience=0.7,
            modality="audio",
            presentation=AudioPresentation(label="warm tone"),
        ),
        ExperienceObject(
            id="round_circle",
            temporal_extent=TemporalExtent(1.9, 3.0, 2),
            features=FeatureVector({
                "sides": 1.0,
                "size": 0.72,
                "hue": 0.58,
                "saturation": 0.82,
                "lightness": 0.76,
                "angularity": 0.0,
                "roundness": 1.0,
                "symmetry": 1.0,
                "novelty": 0.38,
            }),
            salience=0.84,
            modality="visual",
        ),
        ExperienceObject(
            id="small_gap",
            temporal_extent=TemporalExtent(3.35, 4.0, 3),
            features=FeatureVector({
                "sides": 0.0,
                "size": 0.22,
                "hue": 0.0,
                "saturation": 0.0,
                "lightness": 0.84,
                "angularity": 0.04,
                "roundness": 0.08,
                "symmetry": 0.35,
                "novelty": 0.12,
            }),
            salience=0.35,
            modality="visual",
        ),
    ]
)

markdown_table(
    ["id", "start", "end", "modality", "salience", "angularity", "roundness", "novelty"],
    [
        [
            obj.id,
            f"{obj.temporal_extent.start:.1f}",
            f"{obj.temporal_extent.end:.1f}",
            obj.modality,
            f"{obj.salience:.2f}",
            f"{obj.features.value('angularity'):.2f}",
            f"{obj.features.value('roundness'):.2f}",
            f"{obj.features.value('novelty'):.2f}",
        ]
        for obj in sequence.objects
    ],
)
````

### Qualities And Resolved Effects

Authored objects can also start from qualities. A resolver turns those qualities into the flat `salience`, `learning_weight`, and `metadata.affect` fields consumed by the existing model path.

````python
from cave.observation.experience import ExperienceQualities, resolve_experience_object

quality_examples = [
    resolve_experience_object(
        sequence.objects[0],
        ExperienceQualities(magnitude=0.7, novelty=0.4, beauty=0.8, pleasure=0.2),
    ),
    resolve_experience_object(
        sequence.objects[0],
        ExperienceQualities(magnitude=0.7, threat=0.6, pain=0.4),
    ),
]

markdown_table(
    ["example", "salience", "learning", "pain", "pleasure"],
    [
        [
            f"quality_{index}",
            f"{obj.salience:.3f}",
            f"{obj.learning_weight:.3f}",
            f"{obj.metadata['affect']['pain']:.3f}",
            f"{obj.metadata['affect']['pleasure']:.3f}",
        ]
        for index, obj in enumerate(quality_examples)
    ],
)
````

## 2. Subject Configuration

A subject is not just a passive receiver. The relevant API surface is:

- `Sensorium`: which modalities can be sensed.
- `AttentionProfile`: total deployed capacity and channel weights.
- `MemoryParams`: retention, decay, and object-memory lifetime.
- `SubjectiveTopologyParams`: how experience deposits into the topology-like feature plane.
- `ModelParams`: the subject-side configuration passed into the model.

The topology axes below are projections, not literal coordinates. `form` combines angularity, symmetry, and sides. `sensory tone` combines roundness, hue, saturation, and novelty.

````python
from cave.commitments.attention import (
    AttentionChannelCurve,
    AttentionProfile,
    INTERNAL_EXPECTATION_CHANNEL,
)
from cave.commitments.memory import MemoryParams
from cave.commitments.topology import SubjectiveTopologyParams, SubjectiveTopologyPrior
from cave.demonstrations.examples import default_model_params, model_for_sequence
from cave.observation.experience import FeatureProjection
from cave.observation.sensing import Sensorium
from cave.observation.sensing.sensor import FeatureSensor, visual_feature_sensor

sensorium = Sensorium(
    sensors=(
        visual_feature_sensor(),
        FeatureSensor(modality="audio", channel="audio", gain=1.0),
    )
)

base_params = replace(
    default_model_params(),
    memory=MemoryParams(retention=0.78, decay_tau=1.8, max_age=4.0),
    attention=AttentionProfile(
        mode="sine",
        level=0.72,
        amplitude=0.18,
        channel_weights={
            "visual": 0.48,
            "audio": 0.12,
            INTERNAL_EXPECTATION_CHANNEL: 0.40,
        },
        channel_curves={
            "visual": AttentionChannelCurve(mode="sine", level=0.55, amplitude=0.22, phase=0.0),
            "audio": AttentionChannelCurve(mode="sine", level=0.18, amplitude=0.12, phase=math.pi / 2.0, cycles=2.0),
            INTERNAL_EXPECTATION_CHANNEL: AttentionChannelCurve(mode="sine", level=0.45, amplitude=0.26, phase=math.pi),
        },
    ),
    topology=SubjectiveTopologyParams(
        feature_x=FeatureProjection(
            name="form",
            weights={"angularity": 0.5, "symmetry": 0.25, "sides": 0.25},
        ),
        feature_y=FeatureProjection(
            name="sensory tone",
            weights={"roundness": 0.4, "hue": 0.2, "saturation": 0.2, "novelty": 0.2},
        ),
        prior=SubjectiveTopologyPrior(mode="random_wells", strength=0.15, width=0.28, seed=11, well_count=4),
        expectation_deposit_strength=0.16,
        deposit_strength=0.24,
    ),
)

model = model_for_sequence(sequence, params=base_params, vocabulary=vocabulary)
model.sensorium = sensorium

markdown_table(
    ["setting", "value"],
    [
        ["sensor channels", ", ".join(sensor.channel for sensor in sensorium.sensors)],
        ["memory retention", base_params.memory.retention],
        ["attention mode", base_params.attention.mode],
        ["base channel weights", base_params.attention.channel_weights],
        ["topology x", "form = angularity/symmetry/sides"],
        ["topology y", "sensory tone = roundness/hue/saturation/novelty"],
    ],
)
````

## 3. Attention Curves

Attention has a global capacity and a channel distribution. The external sensory channels and the internal expectation channel can move differently over the episode.

````python
def plot_attention_profile(profile: AttentionProfile, duration: float, output: Path) -> Path:
    times = np.linspace(0.0, duration, 160)
    capacities = np.array([profile.value_at(float(t), duration) for t in times])
    channel_names = sorted({*profile.channel_weights, *profile.channel_curves})

    fig, ax = plt.subplots(figsize=(8.6, 3.4))
    ax.plot(times, capacities, color="#111827", linewidth=2.2, label="capacity")
    for channel in channel_names:
        values = np.array([profile.state_at(float(t), duration).channel_weight(channel) for t in times])
        ax.plot(times, values, linewidth=1.8, label=f"{channel} weight")
    ax.set_title("Attention capacity and channel allocation")
    ax.set_xlabel("time")
    ax.set_ylabel("value")
    ax.set_ylim(0.0, 1.05)
    ax.grid(True, alpha=0.25)
    ax.legend(loc="upper right", frameon=False, fontsize=8)
    fig.tight_layout()
    fig.savefig(output, dpi=140)
    plt.close(fig)
    return output

attention_png = plot_attention_profile(base_params.attention, sequence.duration, OUT / "00_attention_curves.png")
show_artifact(attention_png, width=780)
````

## 4. Running The Model

`CaveProducer` runs the model and serializes the result as an `Episode`. The episode is the source-neutral observation contract used by reports, renderers, notebooks, and comparison tools.

````python
from cave.observation.episodes import CaveProducer

producer = CaveProducer(model)
episode = producer.run(dt=0.1)

markdown_table(
    ["field", "value"],
    [
        ["source", episode.source_name],
        ["duration", f"{episode.duration:.2f}"],
        ["inputs", len(episode.inputs)],
        ["observations", len(episode.observations)],
        ["vocabulary", len(episode.vocabulary)],
    ],
)

first = episode.observations[0]
mid = episode.observations[len(episode.observations) // 2]
markdown_table(
    ["observation", "t", "active", "surprise", "learning", "actual norm", "memory norm"],
    [
        ["first", f"{first.t:.2f}", first.active_inputs, f"{first.surprise:.3f}", f"{first.learning_rate:.3f}", f"{np.linalg.norm(first.actual):.3f}", f"{np.linalg.norm(first.memory_state):.3f}"],
        ["middle", f"{mid.t:.2f}", mid.active_inputs, f"{mid.surprise:.3f}", f"{mid.learning_rate:.3f}", f"{np.linalg.norm(mid.actual):.3f}", f"{np.linalg.norm(mid.memory_state):.3f}"],
    ],
)
````

## 5. Inspecting One Timestep

A timestep carries expected input, actual attended input, prediction error, memory state after update, attention metadata, action/value/objective metadata, and topology reconstruction through the structural frame API.

````python
from cave.observation.structural import frame_for_time, structural_state_for_episode

structural = structural_state_for_episode(episode)
frame = frame_for_time(episode, 1.2, structural)
obs = frame.observation

markdown_table(
    ["quantity", "value"],
    [
        ["t", f"{obs.t:.2f}"],
        ["active inputs", obs.active_inputs],
        ["effective external input", f"{obs.metadata['effective_attention']['external_input']:.3f}"],
        ["effective internal expectation", f"{obs.metadata['effective_attention']['internal_expectation']:.3f}"],
        ["surprise", f"{obs.surprise:.3f}"],
        ["learning rate", f"{obs.learning_rate:.3f}"],
        ["memory items visible", len(frame.topology_frame.memory_items)],
        ["topology density mass", f"{np.sum(frame.topology_frame.topology.density):.2f}"],
        ["expected-density mass", f"{np.sum(frame.topology_frame.topology.expected_density):.2f}"],
        ["actual-density mass", f"{np.sum(frame.topology_frame.topology.actual_density):.2f}"],
    ],
)
````

### Internal Expectation Objects

`generate_internal_experiences` is a post-run inspection helper. It turns the episode's expected-input, surprise, and valence readouts into ordinary `ExperienceObject`s with `modality="internal_expectation"`. These generated objects do not feed back into the simulation; they make inward-facing expectation contents easier to inspect with the same vocabulary used for authored inputs.

````python
from cave.observation.experience import generate_internal_experiences

internal_sequence = generate_internal_experiences(episode)
internal_rows = []
for obj in internal_sequence.objects[:5]:
    qualities = obj.metadata["qualities"]
    internal_rows.append([
        obj.id,
        f"{obj.temporal_extent.start:.2f}",
        f"{obj.temporal_extent.end:.2f}",
        obj.modality,
        f"{obj.salience:.3f}",
        f"{obj.learning_weight:.3f}",
        f"{qualities['expectation_magnitude']:.3f}",
        f"{qualities['surprise']:.3f}",
    ])

markdown_table(
    ["id", "start", "end", "channel", "salience", "learning", "expectation", "surprise"],
    internal_rows,
)
````

## 6. Report-Like Views

The view API has a fixed shape:

```text
EpisodeFrame -> View.project(frame) -> ViewState -> renderer.draw(...)
```

Renderers do not advance the model. They read the already-generated episode.

````python
from cave.presentation.renderers.matplotlib_renderer import LayoutSpec, MatplotlibRenderer
from cave.observation.views import (
    PresentationView,
    MemoryLookbackView,
    TimelineView,
    ExpectationActualView,
    CorrectionView,
    SubjectiveTopologyView,
    default_views,
)
from cave.presentation.renderers.topology_surface_renderer import save_topology_state_surface

renderer = MatplotlibRenderer(layout=LayoutSpec(columns=1, dpi=120))

def render_animation(name: str, views, *, dt: float = 0.1, fps: int = 8) -> Path:
    output = OUT / name
    renderer.save_animation(episode, views, output, dt=dt, fps=fps)
    return output

wall_gif = render_animation("01_presentation_wall.gif", [PresentationView()])
memory_gif = render_animation("02_memory_lookback.gif", [MemoryLookbackView()])
timeline_gif = render_animation("03_timeline_attention.gif", [TimelineView()])
expectation_gif = render_animation("04_expectation_actual.gif", [ExpectationActualView()])
correction_gif = render_animation("05_correction.gif", [CorrectionView()])
topology_gif = render_animation("06_subjective_topology.gif", [SubjectiveTopologyView(grid_resolution=42)])

multi_renderer = MatplotlibRenderer(layout=LayoutSpec(columns=2, dpi=120))
multi_gif = OUT / "07_multi_view.gif"
multi_renderer.save_animation(episode, default_views(), multi_gif, dt=0.1, fps=8)

surface_png = OUT / "08_topology_state_surface.png"
save_topology_state_surface(episode, surface_png, resolution=24)

markdown_table(
    ["view", "artifact"],
    [
        ["presentation", wall_gif.relative_to(ROOT)],
        ["memory", memory_gif.relative_to(ROOT)],
        ["timeline", timeline_gif.relative_to(ROOT)],
        ["expectation/actual", expectation_gif.relative_to(ROOT)],
        ["correction", correction_gif.relative_to(ROOT)],
        ["subjective topology", topology_gif.relative_to(ROOT)],
        ["multi-view", multi_gif.relative_to(ROOT)],
        ["topology surface", surface_png.relative_to(ROOT)],
    ],
)
````

### Presentation

The wall view is a projection of active external objects. Wall position is a rendering convention derived from time, not an object field.

````python
show_artifact(wall_gif)
````

### Timeline And Attention

The timeline makes temporal extent and sampled attention visible. These values drive sensing, expectation readout, learning, memory, and topology deposits.

````python
show_artifact(timeline_gif)
````

### Memory

Object memories appear after active intervals end. Their strength follows accumulated attention and decay. The vector memory is also stored on every `EpisodeObservation`.

````python
show_artifact(memory_gif)
````

### Expectation And Actual

Expectation is read before the current update and gated through the `internal_expectation` channel. Actual input is the attended external sensory signal.

````python
show_artifact(expectation_gif)
````

### Correction View

Correction is a diagnostic view of expected, actual, and after-update memory in a feature plane. It is useful, but it is not the same thing as the topology surface.

````python
show_artifact(correction_gif)
````

### Subjective Topology

The topology view shows accumulated density in a projected feature plane. Blue is expected-memory density; green is lived-input density.

````python
show_artifact(topology_gif)
````

### Topology Surface

The surface flattens each topology grid into cells and stacks those cell-density vectors over time. This is the trajectory-like view of the accumulated topology state.

````python
show_artifact(surface_png)
````

### Multi-View

The multi-view animation puts several projections side by side. All panels consume the same episode frames.

````python
show_artifact(multi_gif, width=860)
````

## 7. Configuration Experiments

The same external sequence can produce different subject-side trajectories. Here we keep the sequence fixed and vary only attention configuration.

````python
attention_variants = {
    "balanced_constant": AttentionProfile(
        mode="constant",
        level=0.78,
        channel_weights={"visual": 0.42, "audio": 0.18, INTERNAL_EXPECTATION_CHANNEL: 0.40},
    ),
    "external_heavy": AttentionProfile(
        mode="constant",
        level=0.82,
        channel_weights={"visual": 0.72, "audio": 0.18, INTERNAL_EXPECTATION_CHANNEL: 0.10},
    ),
    "internal_heavy": AttentionProfile(
        mode="constant",
        level=0.82,
        channel_weights={"visual": 0.22, "audio": 0.08, INTERNAL_EXPECTATION_CHANNEL: 0.70},
    ),
    "low_capacity": AttentionProfile(
        mode="constant",
        level=0.24,
        channel_weights={"visual": 0.48, "audio": 0.12, INTERNAL_EXPECTATION_CHANNEL: 0.40},
    ),
}

variant_rows = []
variant_surface_paths = []
for name, attention in attention_variants.items():
    params = replace(base_params, attention=attention)
    variant_model = model_for_sequence(sequence, params=params, vocabulary=vocabulary)
    variant_model.sensorium = sensorium
    variant_episode = CaveProducer(variant_model).run(dt=0.1)
    final = variant_episode.observations[-1]
    surface_path = OUT / f"variant_{name}_surface.png"
    save_topology_state_surface(variant_episode, surface_path, resolution=20)
    variant_surface_paths.append((name, surface_path))
    variant_rows.append([
        name,
        f"{np.linalg.norm(final.memory_state):.3f}",
        f"{np.mean([obs.surprise for obs in variant_episode.observations]):.3f}",
        f"{sum(np.sum(frame.topology.density) for frame in structural_state_for_episode(variant_episode).topology_frames):.1f}",
        surface_path.relative_to(ROOT),
    ])

markdown_table(
    ["variant", "final memory norm", "mean surprise", "total topology mass over frames", "surface"],
    variant_rows,
)
````

The surfaces below share the same world but differ in attention capacity and channel allocation. This is the basic Cave idea: the wall can stay fixed while the subject-side trajectory changes.

````python
for name, path in variant_surface_paths:
    display(Markdown(f"### {name}"))
    show_artifact(path, width=680)
````

## 8. What This Tutorial Establishes

This notebook is the technical version of the walkthrough report:

- external objects are temporal feature-bearing inputs;
- a subject profile determines sensing, attention, memory, expectation, learning, and topology;
- `CaveProducer` serializes the run as an `Episode`;
- views render `EpisodeFrame`s without mutating state;
- topology is an accumulated expected/lived density landscape over feature projections;
- changing subject configuration changes the subjective trajectory even when the external sequence is unchanged.

The next tutorial should move from one trajectory to comparison: same world across subjects, same subject across worlds, embeddings, distance matrices, population dashboards, and report bundles.
