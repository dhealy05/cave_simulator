# Paper 2 Companion: Functional Roles Under Pressure

This notebook is the pedagogical walkthrough for `docs/papers/paper_functional_role_emergence.md`.

It frames the paper as a pressure-to-role operationalization argument: define a role, remove or add the pressure, measure the readout, and state the boundary.

````python
from pathlib import Path
import json
import sys

ROOT = Path.cwd()
if not (ROOT / 'cave').exists():
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

try:
    from IPython.display import Image, Markdown, display
except ModuleNotFoundError:
    class Markdown(str):
        pass
    class Image:
        def __init__(self, filename=None, width=None):
            self.filename = filename
            self.width = width
        def __repr__(self):
            return f'Image({self.filename!r}, width={self.width!r})'
    def display(value):
        print(value)


def show_image(path, width=None):
    path = ROOT / path
    if path.exists():
        display(Image(filename=str(path), width=width))
    else:
        display(Markdown(f'**Missing artifact:** `{path.relative_to(ROOT)}`'))

def md_table(headers, rows):
    header = '| ' + ' | '.join(headers) + ' |'
    sep = '| ' + ' | '.join(['---'] * len(headers)) + ' |'
    body = ['| ' + ' | '.join(str(cell) for cell in row) + ' |' for row in rows]
    display(Markdown('\n'.join([header, sep, *body])))
````

## 1. Pressure-First Hypothesis

The paper is strongest when it avoids claiming consciousness and instead claims operational role readouts under pressure.

````python
role_map = [
    ['limited access', 'selection-like filtering', 'diagnostic preservation under bottleneck'],
    ['temporal recurrence', 'memory-like retention', 'state carries information across delay'],
    ['delayed consequence', 'expectation-like anticipation', 'future outcome decodable before outcome'],
    ['preference/value', 'value-shaped retention', 'valued state retained over neutral/frequent state'],
    ['future coupling control', 'regulation-like agency', 'attention/exposure changes before consequence'],
    ['repeated trajectories', 'topology-like geometry', 'latent/state trajectories separate by role-relevant structure'],
]
md_table(['pressure', 'role readout', 'evidence pattern'], role_map)
````

claim -> manipulation/control -> metric -> visualization -> interpretation -> boundary

## 4. Role Recovery Matrix

This cell computes the compact role-recovery matrix directly from package code. It may take a little time because it runs the checks.

````python
from cave.pressure.tests.role_recovery_matrix import check_role_recovery_matrix

role_result = check_role_recovery_matrix(dt=1.0)
matrix = role_result['matrix']
rows = []
for role, status in matrix.items():
    margins = status.get('margins', {})
    primary = next(iter(margins.items())) if margins else ('', '')
    rows.append([role, status['status'], primary[0], primary[1]])
md_table(['role', 'status', 'example margin', 'value'], rows)
````

## 5. Role Sections

For the paper, each role should get the same treatment: what pressure should force it, what control removes it, what metric measures it, and what caveat remains.

````python
role_sections = [
    ['expectation', 'remove expectation/memory or temporal structure', 'surprise drop, future-outcome readout', 'strong'],
    ['selection', 'remove bottleneck or diagnostic pressure', 'diagnostic workspace/cue concentration', 'weaker; attention-like, not full attention'],
    ['value retention', 'remove preference/value pressure', 'valued focus and hidden value decodability', 'stronger than selection'],
    ['regulation', 'remove future coupling control or reset hidden state', 'exposure/attention contrast before outcome', 'strong in evolved exposure'],
    ['topology geometry', 'compare latent spread/separation against controls', 'value separation / topology mass / trajectory clusters', 'proxy in weak substrates'],
]
md_table(['role', 'control/manipulation', 'metric', 'current read'], role_sections)
````

## 6. Evolved Recurrent Subject

The evolved subject is the most important substrate for Paper 2 because it is not handed Cave's named roles. It has hidden state plus exposure control and is selected for utility in a delayed cue/outcome world.

````python
display(Markdown('''
```text
h_t = tanh(W_x obs_t + W_h h_{t-1} + b_h)
exposure_t = sigmoid(W_a h_t + b_a)

utility_t = exposure_t * outcome_value_t
            - exposure_cost * exposure_t
            - change_cost * abs(exposure_t - exposure_{t-1})
```

The subject is not trained with a prediction loss, explicit memory objective, topology layer, or named expectation variable.
'''))
````

````python
sweep_paths = [
    ROOT / 'out/report_suites/result-ladder/emergence/evolved-exposure-sweep/checks.json',
    ROOT / 'out/report_suites/result-ladder/emergence/evolved-roles-sweep/checks.json',
]
for path in sweep_paths:
    if path.exists():
        data = json.loads(path.read_text())
        display(Markdown(f'### `{path.relative_to(ROOT)}`'))
        display(Markdown(f"checks: **{'pass' if data.get('ok') else 'fail'}**"))
        for item in data.get('extra', []):
            display(Markdown(f"- `{item.get('id')}`: {'pass' if item.get('ok') else 'fail'}"))
    else:
        display(Markdown(f'**Missing:** `{path.relative_to(ROOT)}`'))
````

## 9. What Is Strong vs Weak

This is the section that should make the paper credible. It should explicitly separate robust readouts from fragile ones.

````python
md_table(
    ['claim strength', 'readouts'],
    [
        ['stronger', 'expectation-like future-state readout; exposure regulation; value-shaped latent geometry'],
        ['moderate', 'topology-like separation as a latent geometry proxy'],
        ['weaker', 'selection-like cue sensitivity; not yet full attention'],
        ['boundary', 'operational roles under pressure, not evidence of consciousness'],
    ],
)
````

## 10. Paper 2 Boundary

The paper should avoid sounding like it proves consciousness. The defensible claim is narrower and stronger:

> Several consciousness-adjacent functional roles can be operationalized as subject-side trajectory readouts, destroyed by matched controls, and in some weak substrates selected under specific pressure.

That is enough.
