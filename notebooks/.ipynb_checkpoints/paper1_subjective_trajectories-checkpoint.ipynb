# Paper 1 Companion: Subjective Trajectories

This notebook is the pedagogical walkthrough for `docs/papers/paper_subjective_trajectories.md`.

It does not define new experiments. It walks through existing Cave objects, episodes, views, reports, and curated artifacts.

````python
from pathlib import Path
import json
import sys

ROOT = Path.cwd()
if not (ROOT / 'cave').exists():
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

try:
    from IPython.display import Image, Markdown, display
except ModuleNotFoundError:
    class Markdown(str):
        pass
    class Image:
        def __init__(self, filename=None, width=None):
            self.filename = filename
            self.width = width
        def __repr__(self):
            return f'Image({self.filename!r}, width={self.width!r})'
    def display(value):
        print(value)


def show_image(path, width=None):
    path = ROOT / path
    if path.exists():
        display(Image(filename=str(path), width=width))
    else:
        display(Markdown(f'**Missing artifact:** `{path.relative_to(ROOT)}`'))

def md_table(headers, rows):
    header = '| ' + ' | '.join(headers) + ' |'
    sep = '| ' + ' | '.join(['---'] * len(headers)) + ' |'
    body = ['| ' + ' | '.join(str(cell) for cell in row) + ' |' for row in rows]
    display(Markdown('\n'.join([header, sep, *body])))
````

## 1. The Cave Move

Cave treats experience as a run, not as a picture. External objects pass through a subject profile and become an inspectable subject-side trajectory.

```text
external sequence + subject profile + timestep updates
-> Episode observations
-> views / signatures / reports / population comparisons
```

## 2. Experience Objects

An `ExperienceObject` is the primitive external unit: temporal extent, feature vector, salience, modality, learning weight, and optional presentation. A triangle is not first a screen coordinate; it is a feature object over time.

````python
from cave.demonstrations.examples import demo_model

model = demo_model(seed=7)
sequence = model.sequence
rows = []
for obj in sequence.objects[:4]:
    rows.append([
        obj.id,
        f'{obj.temporal_extent.start:.1f} - {obj.temporal_extent.end:.1f}',
        obj.modality,
        f'{obj.salience:.2f}',
        ', '.join(list(obj.features.values.keys())[:5]) + ('...' if len(obj.features.values) > 5 else ''),
    ])
md_table(['id', 'time', 'modality', 'salience', 'feature keys'], rows)
````

## 3. From Run To Episode

The `Episode` is the common artifact. It is what lets native Cave, GPT-2, conversation, and evolved-subject runs share report and view machinery.

````python
from cave.observation.episodes import CaveProducer

episode = CaveProducer(model).run(dt=0.2)
md_table(
    ['field', 'value'],
    [
        ['source_name', episode.source_name],
        ['inputs', len(episode.inputs)],
        ['observations', len(episode.observations)],
        ['duration', f'{episode.duration:.2f}'],
        ['vocabulary size', len(episode.vocabulary)],
    ],
)
````

## 4. One Timestep

A single observation keeps expected input, actual attended input, prediction error, memory-after-update, attention, surprise, and learning rate separate. This separation is the core of the inspectable trajectory.

````python
obs = episode.observations[len(episode.observations) // 2]
rows = []
for name, expected, actual, error, memory in zip(
    episode.vocabulary,
    obs.expected,
    obs.actual,
    obs.error,
    obs.memory_state,
):
    rows.append([name, f'{expected:.3f}', f'{actual:.3f}', f'{error:.3f}', f'{memory:.3f}'])
display(Markdown(f'**t = {obs.t:.2f}; attention = {obs.attention:.2f}; surprise = {obs.surprise:.3f}; learning_rate = {obs.learning_rate:.3f}**'))
md_table(['feature', 'expected', 'actual', 'error', 'memory after'], rows[:10])
````
